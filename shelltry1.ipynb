```python
import optuna

X_train, X_valid, y_train, y_valid = train_test_split(data, labels, test_size=0.2, random_state=42)
# Objective function for Optuna
def objective(trial):
    # Define hyperparameter search space
    n_estimators = trial.suggest_int("n_estimators", 50, 500)
    max_depth = trial.suggest_int("max_depth", 5, 30)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 20)
    max_features = trial.suggest_categorical("max_features", [None, "sqrt", "log2"])

    # Model
    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        random_state=42,
        n_jobs=-1
    )

    # Train and evaluate
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)

    # Use mean_absolute_percentage_error as the score
    score = mean_absolute_percentage_error(y_valid, y_pred)
    return score  # Optuna minimizes the objective

# Run the study
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=250)

# Print the best result
print("Best trial:")
print(study.best_trial)
print("Best parameters:", study.best_params)
```

```python
import optuna
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, mean_absolute_percentage_error
import numpy as np

def objective(trial):
    # Step 1: Use Optuna to suggest base hyperparameters
    n_estimator = trial.suggest_int('n_estimators', 50, 500, step=50)
    max_depth = trial.suggest_int('max_depth', 5, 30)
    min_sample_leaf = trial.suggest_int('min_samples_leaf', 1, 20)
    min_sample_split = trial.suggest_int('min_samples_split', 2, 20)
    max_features = trial.suggest_categorical('max_features', [None, 'sqrt', 'log2'])
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])

    # Step 2: Define a grid around Optuna's suggestion
    param_grid = {
        'n_estimators': [max(50, n_estimator - 50), n_estimator, min(500, n_estimator + 50)],
        'max_depth': [max(1, max_depth - 5), max_depth, max_depth + 5],
        'min_samples_leaf': [max(1, min_sample_leaf - 1), min_sample_leaf, min_sample_leaf + 1],
        'min_samples_split': [max(2, min_sample_split - 1), min_sample_split, min_sample_split + 1],
        'max_features': [max_features],
        'bootstrap': [bootstrap]
    }

    # Step 3: GridSearchCV using MAPE as scoring (negated because lower is better)
    grid_search = GridSearchCV(
        estimator=RandomForestRegressor(random_state=42, n_jobs=-1),
        param_grid=param_grid,
        scoring=make_scorer(mean_absolute_percentage_error, greater_is_better=False),
        cv=5,
        n_jobs=-1
    )

    grid_search.fit(X_train, y_train)

    return -grid_search.best_score_  # Return positive MAPE so Optuna minimizes it

# Step 4: Run Optuna
study2 = optuna.create_study(direction="minimize")
study2.optimize(objective, n_trials=200)

# Step 5: Best result
print("Best trial:")
print(study2.best_trial)
print("Best parameters:", study2.best_params)
```

```python
from sklearn.model_selection import cross_val_score
def objective(trial):
    # Hyperparameters for GradientBoostingRegressor
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 300),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'max_features': trial.suggest_categorical('max_features', [None, 'sqrt', 'log2'])
    }

    base_model = GradientBoostingRegressor(**params)
    model = MultiOutputRegressor(base_model)

    # Cross-validation
    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    # Mean of negative MAPE (since lower is better)
    scores = cross_val_score(
        model, X_train, y_train,
        scoring=make_scorer(mean_absolute_percentage_error, greater_is_better=False),
        cv=kf,
        n_jobs=-1
    )

    return -np.mean(scores)  # Optuna minimizes this
```

```python
study_gbr = optuna.create_study(direction="minimize")
study_gbr.optimize(objective, n_trials=100)
```
